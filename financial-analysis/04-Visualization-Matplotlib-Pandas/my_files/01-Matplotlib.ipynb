```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# Subplots
plt.subplot(1,2,1)  # row, col, num
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# Object oriented method
fig1 = plt.figure()
ax1 = fig1.add_axes([0.1,0.1,0.8,0.8])
x1 = np.array(x)
y1 = np.array(y)
ax1.plot(x,y)
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_title('Set Title')
plt.show()
```

```python
fig2 = plt.figure()
ax2 = fig2.add_axes([0.1,0.1,0.8,0.8])
ax3 = fig2.add_axes([0.2,0.5,0.4,0.3])
ax2.plot(x,y)
ax3.plot(y,x)
ax2.set_title('LARGER PLOT')
ax3.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
    
axes[0].plot(x,y,'*')
axes[0].set_title('First')

axes[1].plot(y,x,'*')
axes[1].set_title('Second')

plt.show()
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(6,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_first_plot.png',dpi=400)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x^2')
ax.plot(x,x**3,label='x^3')

ax.legend(loc='best')

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF00FF',lw=3,alpha=0.5,ls='--',
        marker='*',markersize=20,markerfacecolor='y',
       markeredgewidth=3,markeredgecolor='k')  # RGB HEX code

ax.plot(x,x**3,'c-.o')

ax.set_xlim([0,4])
ax.set_ylim([0,80])
```
