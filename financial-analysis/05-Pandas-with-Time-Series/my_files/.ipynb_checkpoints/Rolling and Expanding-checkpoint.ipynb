```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../time_data/walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
df.rolling(window=20).mean()['Open'].plot(figsize=(16,6))
```

```python
df['Close SMA30'] = df.rolling(window=30).mean()['Close']
```

```python
df['Close'].plot(figsize=(16,6))
df['Close SMA30'].plot()
plt.legend()
```

```python
df[['Close', 'Close SMA30']].plot(figsize=(16,6))  # Auto legend
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))  # Inverse SMA
```

```python
# Bollinger bands

# Close 20 MA
df['Close SMA20'] = df['Close'].rolling(window=20).mean()

# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close SMA20'] + 2*(df['Close'].rolling(window=20).std())

# Lower = 20MA - 2*std(20)
df['Lower'] = df['Close SMA20'] - 2*(df['Close'].rolling(window=20).std())

# Plot
df[['Close','Close SMA20','Upper','Lower']].tail(200).plot(figsize=(16,6))
```
